# Mapping Columns

```python
columns = ['Gender', 'Hormonal Changes', 'Family History', 'Race/Ethnicity',
                     'Body Weight', 'Calcium Intake', 'Vitamin D Intake', 'Physical Activity',
                     'Smoking', 'Alcohol Consumption', 'Medical Conditions', 'Medications',
                     'Prior Fractures']

df['Gender_Mapped'] = df['Gender'].map({'Male': 0, 'Female': 1})

df['Hormonal_Changes_Mapped'] = df['Hormonal Changes'].map({'Postmenopausal': 0, 'Normal': 1})

df['Family_History_Mapped'] = df['Family History'].map({'No': 0, 'Yes': 1})

df['Race/Ethnicity_Mapped'] = df['Race/Ethnicity'].map({'African': 0, 'Asian': 1, 'Caucasian' : 2})

df['Body_Weight_Mapped'] = df['Body Weight'].map({'Normal': 0, 'Underweight': 1})

df['Calcium_Intake_Mapped'] = df['Calcium Intake'].map({'Adequate': 0, 'Low': 1})

df['Vitamin_D_Intake_Mapped'] = df['Vitamin D Intake'].map({'Insufficient': 0, 'Sufficient': 1})

df['Physical_Activity_Mapped'] = df['Physical Activity'].map({'Active': 0, 'Sedentary': 1})

df['Smoking_Mapped'] = df['Smoking'].map({'Yes': 0, 'No': 1})

df['Alcohol_Consumption_Mapped'] = df['Alcohol Consumption'].map({'Moderate': 0, 'Never': 1})

df['Medical_Conditions_Mapped'] = df['Medical Conditions'].map({'Rheumatoid Arthritis': 0, 'Hyperthyroidism': 1})

df['Medications_Mapped'] = df['Medications'].map({'Corticosteroids': 0, 'Unknown': 1})

df['Prior_Fractures'] = df['Prior Fractures'].map({'No': 0, 'Yes': 1})
```

```python
data = pd.concat([df['Gender_Mapped'],df['Hormonal_Changes_Mapped'],df['Family_History_Mapped'],df['Race/Ethnicity_Mapped'],df['Body_Weight_Mapped'],
                    df['Calcium_Intake_Mapped'],df['Vitamin_D_Intake_Mapped'],df['Physical_Activity_Mapped'],df['Smoking_Mapped'],
                     df['Alcohol_Consumption_Mapped'],df['Medical_Conditions_Mapped'],df['Medications_Mapped'],df['Prior_Fractures']], axis = 1)
```

```python
data

```

```python
combined_df = pd.concat([data,df['Age'], df['Osteoporosis']], axis=1)
combined_df = combined_df.dropna()
```

```python
x = combined_df.drop('Osteoporosis', axis=1)
y = combined_df['Osteoporosis']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size = 0.8, random_state = 42)
```

```python
from sklearn.svm import SVC
```

```python
svcm = SVC(kernel = 'linear')
```

```python
svcm.fit(x_train, y_train)
```

```python
y_pred = svcm.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, y_pred)
print('Accuracy Score : {:.2f}%'.format(acc * 100))
```

```python
from sklearn.metrics import precision_recall_curve, auc, precision_score, roc_curve, roc_auc_score
from sklearn.metrics import confusion_matrix, classification_report
```

```python
from sklearn import metrics
print('', metrics.classification_report(y_test, y_pred))
```

```python
print('',metrics.confusion_matrix(y_test,y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(5, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="flare", xticklabels=['Predicted 0', 'Predicted 1'], yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
auc_roc = roc_auc_score(y_test, y_pred)
print("AUC-ROC Score:", auc_roc)

precision, recall, _ = precision_recall_curve(y_test, y_pred)
auc_prc = auc(recall, precision)
print("AUC-PRC Score:", auc_prc)
```

```python
fpr, tpr, _ = roc_curve(y_test, y_pred)
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(fpr, tpr, label=f'AUC-ROC = {auc_roc:.2f}')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 2)
plt.plot(recall, precision, label=f'AUC-PRC = {auc_prc:.2f}')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc='lower left')
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
cv_scores1 = cross_val_score(svcm, x, y, cv=5, scoring='accuracy')

print("Cross-Validation Scores (Before Tuning):", cv_scores1)
print("Mean CV Score (Before Tuning):", np.mean(cv_scores1))
print("Standard Deviation of CV Scores (Before Tuning):", np.std(cv_scores1))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'C': [0.5, 1, 5, 10, 100],
    'gamma': [0.05, 0.1, 0.5, 1, 10],
    'kernel': ['rbf', 'linear']
}

grid_search = GridSearchCV(svcm, param_grid, refit=True, verbose=2)

grid_search.fit(x_train,y_train)

print('Best Parameters :', grid_search.best_params_)
print('Best Estimator :',grid_search.best_estimator_)

best_model = grid_search.best_estimator_
y_pred = best_model.predict(x_test)


accuracy1 = accuracy_score(y_test, y_pred)
print(f"Test Set Accuracy of Best Model: {accuracy1:.2f}")
```

```python
cv_scores2 = cross_val_score(best_model, x, y, cv=5, scoring='accuracy')

print("Cross-Validation Scores (After Tuning):", cv_scores2)
print("Mean CV Score (After Tuning):", np.mean(cv_scores2))
print("Standard Deviation of CV Scores (After Tuning):", np.std(cv_scores2))
```
